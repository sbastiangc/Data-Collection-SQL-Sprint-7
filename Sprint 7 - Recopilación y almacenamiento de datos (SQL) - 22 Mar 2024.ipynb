### Prueba de Hipótesis

**Hipótesis:** 

**H0:** La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare **NO cambia los sábados lluviosos**

**H1:** La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare **SI cambia los sábados lluviosos**


```python
# creo una tabla dinámica para poder interpretar mejor los resultados
loop_to_ohare.pivot_table(index= 'weather_conditions', values = "duration_seconds", aggfunc =['mean', 'var']).sort_values(('mean',"duration_seconds"), ascending = False)
```

```python
#creo good y bad para analizar la duración de viajes bajo cada condición
Bad = loop_to_ohare[loop_to_ohare['weather_conditions']== 'Bad']
Good = loop_to_ohare[loop_to_ohare['weather_conditions']== 'Good']

# Análisis de Varianza de cada condicion
print(st.levene(Bad['duration_seconds'], Good['duration_seconds']))

# Prueba las hipótesis
alpha = 0.05

# uso t-student para comparar dos muestras con varianzas iguales
results = st.ttest_ind(Bad['duration_seconds'],Good['duration_seconds'],equal_var= True)

print('p-value:', results.pvalue)

# creo un if para el resultado 
if results.pvalue < alpha:
    print("Rechazamos H0")      #si el p-value es menor a 5% rechazamos la Hipotesis nula
else:
    print("No rechazamos H0")   #si el p-value es mayor a 5% NO rechazamos la Hipotesis nula
    
```

**Levene:** No hay diferencia significativa entre las varianzas.

**Rechazamos la Hipótesis Nula:**  La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare **NO cambia los sábados lluviosos**

Si lo vemos gráficamente podemos evidenciar que las condiciones buenas y malas poseen distribuciones significativamente diferentes.

```python
# creo una línea para malas condiciones con kind=kde: para graficar la distribución
Bad['duration_seconds'].plot(kind='kde',
                figsize = (12,6),
                xlim=(0,8000),             
                color= "#FE7A36",
                alpha =0.8,
                title = "Sábados: Duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare"          
                )

# creo una línea para buenas condiciones con kind=kde: para graficar la distribución
Good['duration_seconds'].plot(kind='kde',
                color="steelblue",
                alpha =0.8,
                )

#agrego una legenda a la gráfica, un título al eje x y y
plt.legend(['Malas', 'Buenas'], title='Condiciones climaticas')
plt.xlabel('Duración en segundos')
plt.ylabel('Densidad')
plt.show()
```

**Observaciones:** **Es razonable pensar que las condiciones climáticas afecten el tránsito y la movilidad en la cuidad:** por ejemplo podemos notar que los viajes al aeropuerto duran más en los dias lluviosos, esto puede deberse a embotellamientos, menor visibilidad y reducciones de velocidad. Por otra parte, los dias de buen tiempo tienen más viajes que suelen ser de menor duración.
